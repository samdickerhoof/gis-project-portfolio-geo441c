## **Reflection on data management function in ArcGIS Pro**

- **Write a 100-word paragraph to describe the relationship between**
  - **`Geodatabase`**
  - **`Feature Dataset`**
  - **`Feature Classes`**
  - I can refer to **Page 9** of [assignment guide](https://miamioh.instructure.com/courses/246872/files/folder/week-02-resources?preview=38195670) or other resources online, such as
    - [Fundamentals of geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)

- A geodatabase is basically the main container that holds different kinds of geographic data in one organized place.

- Inside the geodatabase, there are feature datasets, which appear to be folders that group related data together, especially when on the same coordinate system.

- Within those feature datasets are feature classes, which are the actual data layers made up of points, lines, or polygons. They can represent roads, buildings, rivers, etc.

- All of these tie in to make data structured and easy to manage when all correlated together.
